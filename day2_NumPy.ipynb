2. Linear ALgebric Operations

```python
A = np.array([[1,2],
              [3,4]])

B = np.array([[5,6],
              [7,8]])
```

```python
print(np.dot(A,B)) #matrix multiplication
```

```python
A_Trans = A.T #making any vector/matrix transposed
print(A_Trans) 
```

```python
A_Inv = np.linalg.inv(A) 
print(A_Inv)
```

```python
A_Det = np.linalg.det(A) #finding determinant of any matrix
print(A_Det)
```

```python
A_norm = np.linalg.norm(A)
print(A_norm)
```

```python
arr = np.arange(1, 13) 
print(arr)
arr_reshaped = arr.reshape((3,4)) #changing shape of any array
print(arr_reshaped)
arr_flat = arr_reshaped.ravel() #flattens any array
print(arr_flat)
```
